```python
N = 200
appleWidth = np.random.normal(6, 1, N)
appleHeigth = np.random.normal(6, 1, N)

bananasWidth = np.random.normal(4, 1, N)
bananasHeigth = np.random.normal(8, 1, N)
```

```python
plt.scatter(bananasWidth, bananasHeigth, c='r', label='banana')
plt.scatter(appleWidth, appleHeigth, c='orange', label='apple')
plt.xlabel('Width')
plt.ylabel('Heigth')
plt.legend()
```
